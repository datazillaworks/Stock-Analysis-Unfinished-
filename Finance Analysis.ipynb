## Get Dataframe from CSV

```python
## Get Dataframe from CSV
def read_csv(tickers,Path):
    try:
        df = pd.read_csv(Path + tickers + '.csv', index_col=0)
    except FileNotFoundError as ex:
        Print('File Not Found')
    else:
        return df
```

## Merge Dataframe by Column Name

```python
def get_tickers(Path):
    files = [x for x in listdir(Path) if isfile(join(Path,x))]
    tickers = [os.path.splitext(x)[0] for x in files]
    return tickers
```

```python
def merge_df_by_column(col_name,Path):
    mult_df = pd.DataFrame()
    tickers = get_tickers(Path)
    for x in tickers:
        df = read_csv(x,Path)
        mult_df[x] = df.loc[df.index][col_name]
        
    return mult_df
```

## Finance Plotting

```python
def plot_stock(df,ticker):
    fig = go.Figure()
    
    stock_price = go.Scatter(x=df.index, y = df['Close'], 
                            line = dict(color = 'black', width = 1), 
                            name = 'Closing Price')
    ma10 = go.Scatter(x=df.index, y = df['S_Middle_band'], 
                            line = dict(color = 'rgba(0, 0, 250, 0.75)', width = 1), 
                            name = 'MA10')
    ma20 = go.Scatter(x=df.index, y = df['M_Middle_band'], 
                            line = dict(color = 'rgba(0, 0, 250, 0.75)', width = 1), 
                            name = 'MA20')
    ma50 = go.Scatter(x=df.index, y = df['L_Middle_band'], 
                            line = dict(color = 'rgba(0, 0, 250, 0.75)', width = 1), 
                            name = 'MA50')

    fig.add_trace(stock_price)
    fig.add_trace(ma10)
    fig.add_trace(ma20)
    fig.add_trace(ma50)

    
    fig.update_xaxes(title='Date', rangeslider_visible = True)
    fig.update_yaxes(title='Price')
    
    fig.update_layout(title = ticker + ' Price', height = 600, width = 800, 
                      showlegend = True)
    fig.show()
```
